### Map reduce

```python
%%file wordcount.py

from mrjob.job import MRJob 

class MRdateCount(MRJob):
    
   
    def mapper(self, _, date):
         
        yield (date, 1)

   
    def reducer(self, key, values):
        yield (key, sum(values))
        
if __name__ == "__main__":
    MRdateCount.run()
```
